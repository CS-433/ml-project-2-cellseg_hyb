```python
import skimage.io as skio
import skimage
import glob
import matplotlib.pyplot as plt
from sklearn.metrics import jaccard_score
import numpy as np
from bench_filter import Bench_Filter
from sklearn.model_selection import train_test_split
import random
import skimage.morphology as skm
```

```python
from multiprocessing import cpu_count
cpus = cpu_count()
print(f'Number of CPU cores available : {cpus}')
```

```python
DATASET = ['Fluo-N2DL-HeLa','PhC-C2DH-U373']
idx = 0

IM_PATH = sorted(glob.glob(f'data/{DATASET[idx]}/01/*.tif') + glob.glob(f'data/{DATASET[idx]}/02/*.tif'))
TG_PATH = sorted(glob.glob(f'data/{DATASET[idx]}/01_ST/SEG/*.tif') + glob.glob(f'data/{DATASET[idx]}/02_ST/SEG/*.tif'))
```

```python
img, tg = [], []
for im_path, tg_path in zip(IM_PATH,TG_PATH) :
    im = skio.imread(im_path,plugin='pil')
    im = skimage.img_as_float(im)
    im = (im -im.min())/(im.max() - im.min() + 1e-6)
    img.append(im)
    tg.append(skio.imread(tg_path,plugin='pil') > 0)
```

```python
#To avoid memory storage error
#img = img[:int(len(img)/4)] #Divide by 4 for dataset 0 and 2 for dataset 1
#tg = tg[:int(len(tg)/4)]
```

```python
X_train, X_test, y_train, y_test = train_test_split(img,tg,test_size=0.25,shuffle=True,random_state=42)
len(X_train)
```

```python
BF = Bench_Filter(n_estimators=cpus-1, cpus=cpus-1) #Leave 1 cpu to avoid computer crash
BF.add_identity()
BF.add_gaussian(0.5)
BF.add_gaussian(1)
BF.add_gaussian(5) #Progression en sqrt(2)
BF.add_median(skm.disk(3))
BF.add_median()
BF.add_sobel()
BF.add_farid()
BF.add_prewitt()
BF.add_roberts()
BF.add_scharr()
```

```python
BF.fit(X_train, y_train)
```

```python
y_pred = BF.predict(X_test)
```

```python
results = []
for i, pred in enumerate(y_pred):
    results.append(jaccard_score(y_test[i],pred,average='micro'))
plt.hist(results,bins=100,range=(0,1))
plt.title(f'IoU of random forest on {DATASET[idx]}')
plt.xlabel('IoU score')
plt.ylabel('Frequency')
```

```python
n_im = random.randrange(len(y_test))
target = y_test[n_im]
seg = skimage.img_as_bool(BF.predict([X_test[n_im]])[0])
score = jaccard_score(target,seg,average='micro')

fig, ax = plt.subplots(2,2,figsize=(15,10))
ax[0,0].imshow(target,cmap='gray')
ax[0,0].set_title('Target')
ax[0,1].imshow(seg,cmap='gray')
ax[0,1].set_title(f'Score prediction : {score:.2f}') 
ax[1,0].imshow(target & ~seg,cmap='gray')
ax[1,0].set_title('False Negative')
ax[1,1].imshow(~target & seg,cmap='gray') 
ax[1,1].set_title('False Positive')
```
